## DOMAIN ANALYSIS:



1. **Erythema:** Redness of the skin, often indicating inflammation.

2. **Scaling:** The presence of scales on the skin, which can be associated with various skin conditions.

3. **Definite Borders:** Clear and well-defined borders of skin lesions, which can be important in diagnosing certain skin disorders.

4. **Itching:** An indication of discomfort or irritation associated with the skin.

5. **Koebner Phenomenon:** Development of skin lesions in areas of trauma or injury.

6. **Polygonal Papules:** Small, raised, polygon-shaped skin lesions.

7. **Follicular Papules:** Lesions affecting hair follicles.

8. **Oral Mucosal Involvement:** Involvement of the mucous membranes inside the mouth.

9. **Knee and Elbow Involvement:** Presence of symptoms or lesions on the knees and elbows.

10. **Scalp Involvement:** Involvement of the scalp, which can be indicative of certain skin conditions.

11. **Family History:** The presence of similar skin conditions in family members.

12. **Melanin Incontinence:** Loss of melanin in the skin.

13. **Eosinophils in the Infiltrate:** Presence of a type of white blood cell associated with inflammation.

14. **PNL Infiltrate:** Presence of inflammatory cells in the skin.

15. **Fibrosis of the Papillary Dermis:** Thickening and scarring of the connective tissue in the upper layer of the skin.

16. **Exocytosis:** Infiltration of inflammatory cells into the epidermis.

17. **Acanthosis:** Thickening of the epidermis.

18. **Hyperkeratosis:** Thickening of the stratum corneum, the outermost layer of the epidermis.

19. **Parakeratosis:** Retention of nuclei in the stratum corneum.

20. **Clubbing of the Rete Ridges:** Enlargement of the spaces between the skin ridges.

21. **Elongation of the Rete Ridges:** Lengthening of the ridges in the skin.

22. **Thinning of the Suprapapillary Epidermis:** Reduction in thickness of the epidermal layer.

23. **Spongiform Pustule:** A type of pustule characterized by spongiosis.

24. **Munro Microabcess:** Small abscesses in the skin.

25. **Focal Hypergranulosis:** Abnormal thickening of the granular layer of the epidermis.

26. **Disappearance of the Granular Layer:** Loss of the granular layer of the epidermis.

27. **Vacuolisation and Damage of Basal Layer:** Changes in the basal layer of the epidermis.

28. **Spongiosis:** Presence of edema in the epidermis.

29. **Saw-tooth Appearance of Rete Ridges:** Irregular appearance of the skin ridges.

30. **Follicular Horn Plug:** Accumulation of material in hair follicles.

31. **Perifollicular Parakeratosis:** Retention of nuclei around hair follicles.

32. **Inflammatory Mononuclear Infiltrate:** Presence of inflammatory cells in the skin.

33. **Band-Like Infiltrate:** Linear infiltration of inflammatory cells.

34. **Age:** The age of the patient, which can be a crucial factor in dermatological diagnoses.

35. **Class:** The target variable indicating the presence or absence or stages of a particular skin condition.

# KNN-REGRESSOR

```python


from sklearn.neighbors import KNeighborsClassifier  #it is classification task
# create a list to store the error values for each k
error_rate = []

# Will take some time  finding the k value
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)   #pred_i=predicted o/p
    error_rate.append(np.mean(pred_i != y_test)) #ytest---actual o/p
    # if predicted value is not equal to actual value, returns true (which is taken as 1) else false(0). 
    # Adds all the value and takes mean of it. So for each k-value, gets the mean of error.
    #print(np.mean(pred_i != y_test))
```

```python
# Lets plot the finded k-value and error rate

plt.figure(figsize=(10,6))
plt.plot(range(1,11),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# let's fit the data into KNN model and see how well it performs:
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)# Predict 
y_pred = knn.predict(X_test)
```
